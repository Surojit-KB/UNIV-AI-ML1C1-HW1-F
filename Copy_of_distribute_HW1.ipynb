### 1.3 Parse the data

Now, write a function `parse_year(year, yeartextdict)` that returns a dictionary `yearinfo`.
Here `year` should be the integer value of the year, and `yeartextdict` should be the HTML text for the corresponding webpage. 
This function should return a dictionary of the following form:

{'ranking': 32,

 'songurl': ['/wiki/Hot_Boyz_(song)'],
 'titletext': '"Hot Boyz"',

Notice that the `titletext` is the contents of the table cell, and retains the quotes that wikipedia puts on the single.

Here are some issues that you will need to take care of:

````python
"""
Function
--------
parse_year

Inputs
------
year: the year you want the singles for
ytextdixt: a dictionary with keys as integer years and values the downloaded web pages 
    from wikipedia for that year.
   
Returns
-------

a list of dictionaries, each of which corresponds to a single and has the
following data:

Eg:

{'band_singer': ['Brandy', 'Monica'],
  'ranking': 2,
  'song': ['The Boy Is Mine'],
  'songurl': ['/wiki/The_Boy_Is_Mine_(song)'],
  'titletext': '" The Boy Is Mine "',
  'url': ['/wiki/Brandy_Norwood', '/wiki/Monica_(entertainer)']}
  
A dictionary with the following data:
    band_singer: a list of bands/singers who made this single
    song: a list of the titles of songs on this single
    songurl: a list of the same size as song which has urls for the songs on the single 
        (see point 3 above)
    ranking: ranking of the single
    titletext: the contents of the table cell
    band_singer: a list of bands or singers on this single
    url: a list of wikipedia singer/band urls on this single: only put in the part 
        of the url from /wiki onwards
    
"""
#your code here
def parse_year(year,yeartextdict):
  yeardata=[]
  soup=BeautifulSoup(yeartextdict[year])
  rows=soup.find('table',attrs={'class':'wikitable'}).findChildren('tbody')[0].findChildren('tr') #SIMILAR TO ABOVE CODE FOR 1970
  for i in range(1,len(rows)):
    row=rows[i]
    yeardata.append({'song':[link['title'] for link in row.findChildren('td')[-2].findChildren('a')],
                     'titletext':row.findChildren('td')[-2].text.strip(),
                  'songurl':[link['href'] for link in row.findChildren('td')[-2].findChildren('a')],
                  'url':[a_link['href'] for a_link in row.findChildren('td')[-1].findChildren('a')],
                  'ranking':int(row.findChildren()[0].text.strip()),
                  'band_singer':[a_link['title'] for a_link in row.findChildren('td')[-1].findChildren('a')]})
    

    #CONDITIONS FOR BREAKING OFF THE LIST AND CREATING SEPARATE DICTIONARIES FOR EACH ENTRY THAT HAS MULITPLE SONGS OR MULTIPLE ARTISTS
    #TRIED PD.EXPLODE BEFORE BUT WAS NOT GIVING CORRECT RESULTS
    #LOGIC USED WAS IF AN ENTRY HAS MULITPLE BAND SINGER THEN IT WILL MOSTLY HAVE MULTIPLE SINGER URL AND WE CAN CREATE THE ENTRY BASED ON LEN(BAND_SINGER)
    #SIMILAR LOGIC FOR MULTIPLE SONG ENTRIES
    #LOGIC BREAKSDOWN: IF SONG URL OR BAND_SINGER URL IS NOT AVAILABLE. BUT WONT BREAK HERE CUZ WE GOT THE TITLE FROM HREF WHILE GETTING SINGER/SONG URL.
    #HENCE, WE WONT GET THE ENTRIES WHERE SONG OR SINGER URL IS NOT PRESENT. THIS CAVEAT COULD BE OVERCOME BUT WAS NOT ATTEMPTED DUE TO TIME CONSTRAINTS
    if (len(yeardata[-1]['band_singer'])>1)&(not len(yeardata[-1]['songurl'])>1): #CONDITION FOR MULTIPLE SINGERS ONLY
      dic1=yeardata.pop()
      for j in range(len(dic1['band_singer'])): #CREATING SEPARATE ENTRIES ON THE BASIS OF LENGTH OF SINGER LIST
        temp_dic=dic1.copy()
        temp_dic['band_singer']=dic1['band_singer'][j]
        temp_dic['url']=dic1['url'][j]
        yeardata.append(temp_dic)
    elif (not len(yeardata[-1]['band_singer'])>1)&(len(yeardata[-1]['songurl'])>1): #CONDITION FOR MULTIPLE SONGS
      dic1=yeardata.pop()
      for j in range(len(dic1['songurl'])): #CREATING ENTRIES FOR MULTIPLE BASED ON LENGTH OF MULTIPLE SONGS
        temp_dic=dic1.copy()
        temp_dic['songurl']=dic1['songurl'][j]
        temp_dic['song']=dic1['song'][j]
        yeardata.append(temp_dic)
    elif (len(yeardata[-1]['band_singer'])>1)&(len(yeardata[-1]['songurl'])>1): #CONDITION FOR MULTIPLE SINGER AS WELL AS SONG
      dic1=yeardata.pop()
      for j in range(len(dic1['songurl'])): #CREATING ALL POSSIBLE MULTIPLE ENTRIES
        for k in range(len(dic1['band_singer'])):
          temp_dic=dic1.copy()
          temp_dic['songurl']=dic1['songurl'][j]
          temp_dic['song']=dic1['song'][j]
          temp_dic['band_singer']=dic1['band_singer'][k]
          temp_dic['url']=dic1['url'][k]

      
  return yeardata
````

````python
#your code here
yearinfo= {} #GETTING THE ACTUAL DATA
for i in range(1992,2021):
  yearinfo[i]=parse_year(i,yearstext)
````

### 3.1 Scrape the artist's Wikipedia webpages 

Since we have hundreds of artists webpages to scrape, we have created a function which implements caching in order to speed up this process.

The cache object `urlcache` that will avoid redundant HTTP requests (e.g. an artist might have multiple singles on a single year, or be on the list over a span of years). Remember that this function is designed to be used in a loop over years, and then a loop over songs per year. Since network requests are relatively slow, if we have already requested for a singer or band's wikipedia page, caching the results is a smart thing to do.



Notice that we have wrapped the call in _an exception block_. If the request gets an HTTP code different from 200, the cells for that URL will have a value of 1; and if the request completely fails (e.g. no network connection) the cell will have a value of 2. This will allow you to analyse the failed requests.

````python
urlcache={}
def get_page(url):
    if (url not in urlcache) or (urlcache[url]==1) or (urlcache[url]==2):
        time.sleep(1)
        try:
            r = requests.get("http://en.wikipedia.org%s" % url)
            if r.status_code == 200:
                urlcache[url] = r.text
            else:
                urlcache[url] = 1
        except:
            urlcache[url] = 2
    return urlcache[url]
````

Before we apply our function to the dataframe, let us sort `flatframe` by year. This will ensure that we will hit the cache most as singers who show up repeatedly in the rankings will have their information already pulled.

````python
genres_duplicates={'/wiki/Adult_Contemporary_music':'/wiki/Adult_contemporary','/wiki/Adult_contemporary_music':'/wiki/Adult_contemporary',
'/wiki/Afrobeat':'/wiki/Afrobeats',
'/wiki/Alternative_rock':'/wiki/Alternative_Rock',
'/wiki/Avant-garde':'/wiki/Avant-garde_music',
'/wiki/Blues':'/wiki/Blues_music',
'/wiki/Comedy_hip-hop':'/wiki/Comedy_hip_hop',
'/wiki/Contemporary_R%26B':'/wiki/Contemporary_R%26B_music',
'/wiki/Contemporary_folk':'/wiki/Contemporary_folk_music',
'/wiki/Country_Folk':'/wiki/Country_folk',
'/wiki/Dance_pop':'/wiki/Dance-pop',
'/wiki/East_Coast_hip_hop':'/wiki/East_coast_hip_hop',
'/wiki/Electronic_Dance_Music':'/wiki/Electronic_dance_music',
'/wiki/Electronica':'/wiki/Electronica_music',
'/wiki/Emo':'/wiki/Emo_music',
'/wiki/Electropop':'/wiki/Electro-pop',
'/wiki/Folk-pop':'/wiki/Folk_pop',
'/wiki/Funk':'/wiki/Funk_music',
'/wiki/Grime_(music_genre)':'/wiki/Grime_music',
'/wiki/Gangsta_Rap':'/wiki/Gangsta_rap',
'/wiki/Hip_Hop_music': '/wiki/Hip_hop','/wiki/Hip_hop_music':'/wiki/Hip_hop',
'/wiki/Hyphy':'/wiki/Hyphy_music',
'/wiki/Latin_music':'/wiki/Latin_music_(genre)',
'/wiki/West_Coast_hip_hop':'/wiki/West_coast_hip_hop',
'/wiki/Southern_Hip_Hop':'/wiki/Southern_hip_hop',
'/wiki/Ska':'/wiki/Ska_music',
'/wiki/Pop-rock':'/wiki/Pop_rock',
'/wiki/Pop_Music':'/wiki/Pop_music',
'/wiki/Nu_metal':'/wiki/Nu_metal_music',
'/wiki/Hard_Rock':'/wiki/Hard_rock',
'/wiki/Pop_Rock':'/wiki/Pop_rock',
'/wiki/Post-Grunge':'/wiki/Post-grunge',
'/wiki/SoundCloud_rap':'/wiki/Soundcloud_rap'}

def genre_duplicates(genres):
    for i in range(len(genres)):
        if genres[i] in genres_duplicates.keys():
            genres[i]=genres_duplicates[genres[i]]
    return genres
````

````python
#your code here 
# NOW COMES THE  BIG CHALLENGE
# TRIED TO USE THE REGEX PROVIDED, WAS NOT ABLE TO CORRECTLY THE DATA BECAUSE OF MY LACK OF KNOWLDGE OF REGEX
# SUPER MEGA ENTENSIVE ANALYSIS OF THE WIKIPEDIA PAGES GAVE THREE MAJOR INFO ABOUT THE YEAR ACTIVE ENTRY
# WIKIPEDIA PAGES YEARS ACTIVE ENTRY ARE OF THE FOLLOWING TYPES:
# 1) WITH MULTIPLE YEARS ACTIVE, THE ENTRY IS SEPARATED USING EITHER BREAK (<br>) TAG OR <li> TAG. USED REGEX TO FIND THIS TAGS
# 2) SOMETIMES THERE ARE SOME REFERENCES IN THE YEARS ACTIVE WHICH CREATES PROBLEM FOR GETTING THE CORRECT DATE.
#    THESE ARE USUALLY IN THE FORM OF <span> OR <sup> TAG. USED DECOMPOSE TO GET RID OF THESE TAGS
# 3) SOMETIMES THE YEARS ARE SEPARATED USING COMMAS OR SEMICOLONS. USED A SEPARATE CONDITION FOR THESE
# 4) SOME ENTRIES HAVE HIATUS OR REUNION INFO WHICH COULD HAVE BEEN USED BUT WAS NOT DUE TO TIME CONSTRAINTS. THIS WERE REMOVED USING A COMBINATION OF 
#    REGEX MATCHING AND STRING PARTITON.


def years_active(url):
   soup=BeautifulSoup(url,'html.parser')
   try:
      for child1 in soup.find('table',attrs={'class':'infobox'}).findChildren('tr'):
        for child2 in child1.findChildren('th',attrs={'scope':'row'}):
          if (re.search('Years',child2.text)):
            imp_child=child1
            break
   except AttributeError:
      final_year='nan'
      return final_year

    #LOOP FOR BREAK TAG MATCHING AND GETTING THE YEARS ACTIVE INFO
   if (re.search('<br/>',str(imp_child.contents))):

      if (re.search('span',str(imp_child.findChildren('td')[0]))):
        while (re.search('span',str(imp_child.findChildren('td')[0]))):
          imp_child.findChildren('td')[0].span.decompose()          #DECOMPOSING SPAN TAGS IF ANY
        str1=imp_child.findChildren('td')[0].get_text(separator='|').replace('|',"",1)

      elif (re.search('sup',str(imp_child.findChildren('td')[0]))):
        while (re.search('sup',str(imp_child.findChildren('td')[0]))):
          imp_child.findChildren('td')[0].sup.decompose()           #DECOMPOSING SUP TAGS IF ANY
        str1=imp_child.findChildren('td')[0].get_text(separator='|').replace('|',"",1).strip()

      else:
        str1=imp_child.findChildren('td')[0].get_text(separator='|').strip()
      if (re.search('reunion',str1.lower())):                       #REMOVING HIATUS/REUNION INFO
        str1=str1.lower().partition('(')[0]
      elif (re.search('hiatus',str1.lower())):
        str1=str1.lower().partition('(')[0] 

      l1=str1.strip().split('|')                                    #CREATING A LIST OF THE TEXT EXTRACTED
      diff=[]
      for i in range(len(l1)):                                      #LOOP FOR GETTING THE YEARS FROM THE DATA USING REGEX MATCHING
        l1[i]=l1[i].lower().replace('present','2021')               #REPLACING PRESENT WITH 2021 FOR CALCULATING MAX YEAR ACTIVE
        year_list=re.findall(r'\d{4}',l1[i])        
        try:
          y1=int(year_list[0])                                      #GETTING THE DIFFERENCE BETWEEN THE YEARS ACTIVE FOR GETTING THE MAX YEARS ACTIVE
          y2=int(year_list[1])
          diff.append(abs(y2-y1))
        except IndexError:
          pass       
      final_year=l1[diff.index(max(diff))]                          #GETTING THE FINAL MAX YEARS ACTIVE
      return final_year.replace('2021','present')                   #RETURNING THE YEARS ACTIVE

    #LOOP FOR HANDLING THE DATA FROM <li> TAG. INDIVIDUAL LOOPS ARE SIMILAR TO ABOVE CODE.
   elif (re.search('<li>',str(imp_child.contents))):
      if (re.search('span',str(imp_child.findChildren('td')[0]))):
        while (re.search('span',str(imp_child.findChildren('td')[0]))):#SPAN DECOMPOSE
          imp_child.findChildren('td')[0].span.decompose()
        str1=imp_child.findChildren('td')[0].get_text(separator='|').replace('|',"",1).strip()

      elif (re.search('sup',str(imp_child.findChildren('td')[0]))):#SUP DECOMPOSE
        while (re.search('sup',str(imp_child.findChildren('td')[0]))):
          imp_child.findChildren('td')[0].sup.decompose()
        str1=imp_child.findChildren('td')[0].get_text(separator='|').replace('|',"",1).strip()

      else:
        str1=imp_child.findChildren('td')[0].get_text(separator='|').strip()
      if (re.search('reunion',str1.lower())):#REMOVING REUNION/HIATUS
        str1=str1.lower().partition('(')[0]
      elif (re.search('hiatus',str1.lower())):
        str1=str1.lower().partition('(')[0]


      l1=str1.strip().split('|')
      for k,i in enumerate(l1):
        if i=='':
          l1.pop(k)
          pass
      diff=[]

      for i in range(len(l1)):
        l1[i]=l1[i].lower().replace('present','2021')
        year_list=re.findall(r'\d{4}',l1[i])
        try:
          y1=int(year_list[0])
          y2=int(year_list[1])
          diff.append(abs(y2-y1))
        except IndexError:
          pass       
      final_year=l1[diff.index(max(diff))]
      return final_year.replace('2021','present') #GETTING THE MAX YEARS ACTIVE 

    #CODE FOR ALL OTHER CASES INCLUDING COMA AND SEMICOLON SEPARATED VALUES
   else:
     try:
        if (re.search('span',str(imp_child.findChildren('td')[0]))):
          while (re.search('span',str(imp_child.findChildren('td')[0]))):
            imp_child.findChildren('td')[0].span.decompose()
          str1=imp_child.findChildren('td')[0].get_text(separator='|').replace('|',"",1).strip()
        elif (re.search('sup',str(imp_child.findChildren('td')[0]))):
          while (re.search('sup',str(imp_child.findChildren('td')[0]))):
            imp_child.findChildren('td')[0].sup.decompose()
          str1=imp_child.findChildren('td')[0].get_text(separator='|').replace('|',"",1).strip()
        else:
          str1=imp_child.findChildren('td')[0].get_text(separator='|').strip()
        
        str1=str1.lower().replace('present','2021')

        #REUINON/HIATUS HANDLING
        if (re.search('reunion',str1.lower())):
          str1=str1.lower().partition('(')[0]
        elif (re.search('hiatus',str1.lower())):
          str1=str1.lower().partition('(')[0]

        #COMMA/SEMICOLON HANDLING
        if (re.search(',',str1)) or (re.search(';',str1)):
          if re.search(',',str1):
            l1=str1.split(',')
          elif re.search(';',str1):
            l1=str1.split(';')
        else:
          l1=str1.strip().split('|')
        diff=[]
        for i in range(len(l1)):
          l1[i]=l1[i].replace('present','2021')
          year_list=re.findall(r'\d{4}',l1[i])
          try:
            y1=int(year_list[0])
            y2=int(year_list[1])
            diff.append(abs(y2-y1))
          except IndexError:
            pass
        final_year=l1[diff.index(max(diff))]
        if re.search(';',final_year):
            final_year=final_year.split(';')[0]
        return final_year.replace('2021','present')
     except ValueError:
       return str1

      
   return final_year
   #year_set=list(set(re.findall(r'\d{4}',str1)))
   



````

````python
"""
Function
--------
singer_band_info

Inputs
------
url: the url
page_text: the text associated with the url
   
Returns
-------
A dictionary with the following data:
    url: copy the input argument url into this value
    genres: the genres that the band or singer works in
    born: the artist's birthday
    ya: years active variable

Notes
-----
See description above. Also note that some of the genres urls might require a 
bit of care and special handling.
"""

#your code here
def singer_band_info(url):
  soup=BeautifulSoup(urlcache[url],'html.parser')
  #GETTING YEARS ACTIVE
  try:
    y_a=years_active(urlcache[url])
  except UnboundLocalError:
    y_a=np.nan
  except AttributeError:
    y_a=np.nan
  #GETTING DOB DETAILS
  try:
    for i in soup.find_all('th',attrs={'scope':'row'}):
      if re.search('Born',i.text):#REGEX MATCHING FOR BORN
        flag1=True
        break
      else:
        flag1=False
    if flag1:
      bday=soup.find_all('span',attrs={'class':'bday'})[0].text
    else:
      bday=np.nan
  except IndexError:
    bday=np.nan
  except UnboundLocalError:
    bday=np.nan
  #GETTING GENRES
  try:
    genre=get_genres(soup)
  except UnboundLocalError:
    genre=np.nan
  singer_info_dict={'url':url,
                    'genres':genre,
                    'born': bday,
                    'ya':y_a}
  return singer_info_dict#RETURNING DICTIONARY
````
